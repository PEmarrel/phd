```python
import gensim.models
from gensim.test.utils import datapath
from gensim import utils
from gensim.models.callbacks import CallbackAny2Vec
```

```python
class MyCorpus:
    """An iterator that yields sentences (lists of str)."""

    def __iter__(self):
        for line in open('data_base.txt'):
            yield utils.simple_preprocess(line)
```

```python
lr = 0.025
emb_dim = 16
nb_negative = 5
epochs = 10
threshold = 1#Â (0, 1e-5)
min_count=1
window=5

sentences = MyCorpus()
model = gensim.models.Word2Vec(sentences=sentences, alpha=lr, vector_size=emb_dim, min_count=min_count,
                                negative=nb_negative, epochs=1, sample=threshold, seed=1, window=window)
```

```python
print(model.wv['chat'])
print(model.wv.most_similar(positive=['chat'], topn=5))
```

```python
class EpochLogger(CallbackAny2Vec):

    '''Callback to log information about training'''


    def __init__(self):

        self.epoch = 0


    def on_epoch_begin(self, model):

        print("Epoch #{} start".format(self.epoch))


    def on_epoch_end(self, model:gensim.models.Word2Vec):

        print("Epoch #{} end".format(self.epoch))

        self.epoch += 1
```

```python
epoch_logger = EpochLogger()
model.train("data_base.txt", total_examples=len(sentences), epochs=10, callbacks=[epoch_logger])
```

```python
print(model.wv['chat'])
print(model.wv.most_similar(positive=['chat'], topn=5))
```
